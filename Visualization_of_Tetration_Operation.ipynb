# Visualization of Tetration Operation (테트레이션 연산 시각화)

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
max_iter = 100  # 최대 반복 횟수
escape_radius = 10  # 발산 조건, 복소수 크기가 이 값보다 커지면 발산으로 간주
resolution = 500  # 해상도
xmin, xmax = -2, 2  # x 축 범위
ymin, ymax = -2, 2  # y 축 범위

# Create a grid of complex numbers
x = np.linspace(xmin, xmax, resolution)
y = np.linspace(ymin, ymax, resolution)
X, Y = np.meshgrid(x, y)
c_val = X + 1j * Y  # 복소평면의 각 점에 대한 복소수 값

# Initialize divergence map
divergence_map = np.zeros((resolution, resolution), dtype=bool)

# Perform iteration
for i in range(resolution):
    for j in range(resolution):
        z = 0  # 초기값
        for k in range(max_iter):
            z = c_val[i, j] ** z  # 테트레이션 연산
            if np.abs(z) > escape_radius:
                divergence_map[i, j] = True  # 발산으로 처리
                break

# Plot the Mandelbrot set
plt.figure(figsize=(8, 8))
plt.imshow(divergence_map.T, extent=(xmin, xmax, ymin, ymax), cmap='hot', origin='lower')
plt.title('Divergence Map')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.colorbar(label='Divergence')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 매개변수 설정
max_iter = 100  # 최대 반복 횟수
escape_radius = 10  # 발산 조건, 복소수 크기가 이 값보다 커지면 발산으로 간주
resolution = 500  # 해상도
xmin, xmax = -2, 2  # x 축 범위
ymin, ymax = -2, 2  # y 축 범위

# 복소평면에 그리드 생성
x = np.linspace(xmin, xmax, resolution)
y = np.linspace(ymin, ymax, resolution)
X, Y = np.meshgrid(x, y)
c_val = X + 1j * Y  # 각 점에 대한 복소수 값

# 발산 지도 초기화
divergence_map = np.zeros((resolution, resolution), dtype=bool)

# 반복 수행
for i in range(resolution):
    for j in range(resolution):
        z = 0  # 초기값
        for k in range(max_iter):
            z = c_val[i, j] ** z  # 테트레이션 연산
            if np.abs(z) > escape_radius:
                divergence_map[i, j] = True  # 발산으로 처리
                break

# 맨델브로트 집합 시각화
plt.figure(figsize=(8, 8))
plt.imshow(divergence_map.T, extent=(xmin, xmax, ymin, ymax), cmap='hot', origin='lower')
plt.title('발산 지도')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.colorbar(label='발산')
plt.show()
```
